# **Question 1**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np

c = np.array([(50, 50), (50, 100), (150, 255), (150, 150), (255,255)]) # (x, y) coordinates of control points

t1 = np.linspace(0, c[0, 1], c[0, 0] + 1 - 0).astype('uint8') # 0 to 50, 51 elements
print(len(t1))
t2 = np.linspace(c[1, 1] + 1, c[2,1], c[2, 0] - c[1, 0]).astype('uint8') # 101 to 255, 100 elements
print(len(t2))
t3 = np.linspace(c[3, 1] + 1, c[4, 1], c[4, 0] - c[3, 0]).astype('uint8') # 151 to 255, 105 elements
print(len(t3))

transform = np.concatenate((t1, t2), axis=0).astype('uint8') # concatenate arrays
transform = np.concatenate((transform, t3), axis=0).astype('uint8') # concatenate arrays
print(len(transform))

fig, ax = plt.subplots()
ax.plot(transform)
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel('Output, $\mathrm{T}[f(\mathbf{x})]$')
ax.set_xlim(0, 255)
ax.set_ylim(0, 255)
ax.grid(True, which='both') # show grid lines
ax.set_aspect('equal')
plt.savefig('emma_transformation.png')
plt.show()

#print original image and transformed image
fig, ax = plt.subplots(1, 2)

img_orig = cv.imread('emma.jpg', cv.IMREAD_GRAYSCALE) # read image as grayscale
ax[0].imshow(img_orig, cmap='gray')
ax[0].set_title('Original Image')
ax[0].axis('off')

img_transformed = cv.LUT(img_orig, transform)
ax[1].imshow(img_transformed, cmap='gray')
ax[1].set_title('Transformed Image')
ax[1].axis('off')

plt.show()

#Save transformed image
# fig, ax = plt.subplots()
# ax.imshow(img_transformed, cmap='gray')
# ax.axis('off')
# plt.savefig('emma_transformed.png')
```

```python
%matplotlib inline
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
im = np.zeros((6,8), dtype=np.uint8) #An image with a height of 6 pixels, a width of 8 pixels, and 3 color channels (Red, Green, and Blue). The array is initialized with all zeros, creating a black image. Data type is unsigned integer 8-bit
im[4,5] = 150                       #Set the top-left pixel to white
im[4,6] = 200                     #Set the top-left pixel to white
fig, ax = plt.subplots()             #fig = canvas and ax = image
ax.imshow(im, cmap='gray', vmin=0, vmax=255) #Display the image in grayscale with lowest pixel value = 0 and highest pixel value = 255
ax.xaxis.tick_top() #Move the x-axis label to the top
plt.show()  
```
