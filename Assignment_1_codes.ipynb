```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np

c = np.array([(50, 50), (50, 100), (150, 255), (150, 150), (255,255)]) # (x, y) coordinates of control points

t1 = np.linspace(0, c[0, 1], c[0, 0] + 1 - 0).astype('uint8') # 0 to 50, 51 elements
print(len(t1))
t2 = np.linspace(c[0, 1] + 1, c[1, 1], c[1, 0] - c[0, 0]).astype('uint8') # 51 to 100, 0 elements
print(len(t2))
t3 = np.linspace(c[1, 1] + 1, c[2,1], c[2, 0] - c[1, 0]).astype('uint8') # 101 to 255, 100 elements
print(len(t3))
t4 = np.linspace(c[2, 1] - 1, c[3, 1], c[3, 0] - c[2, 0]).astype('uint8') # 254 to 150, 0 elements
print(len(t4))
t5 = np.linspace(c[3, 1] + 1, c[4, 1], c[4, 0] - c[3, 0]).astype('uint8') # 151 to 255, 105 elements
print(len(t5))

transform = np.concatenate((t1, t2), axis=0).astype('uint8') # concatenate arrays
transform = np.concatenate((transform, t3), axis=0).astype('uint8') # concatenate arrays
transform = np.concatenate((transform, t4), axis=0).astype('uint8') # concatenate arrays
transform = np.concatenate((transform, t5), axis=0).astype('uint8') # concatenate arrays
print(len(transform))

fig, ax = plt.subplots()
ax.plot(transform)
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel('Output, $\mathrm{T}[f(\mathbf{x})]$')

ax.set_xlim(0, 255)
ax.set_ylim(0, 255)
ax.grid(True, which='both') # show grid lines
ax.set_aspect('equal')
plt.savefig('transform.png')
plt.show()

img_orig = cv.imread('emma.jpg', cv.IMREAD_GRAYSCALE) # read image as grayscale
cv.namedWindow("Image", cv.WINDOW_AUTOSIZE)
cv.imshow("Image", img_orig)
cv.waitKey(0)
image_transformed = cv.LUT(img_orig, transform)
cv.imshow("Transformed Image", image_transformed)
cv.waitKey(0)
cv.destroyAllWindows()
```
